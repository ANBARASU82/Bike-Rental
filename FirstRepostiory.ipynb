IMPORTING THE DATASET

```python
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/SAL/bikerent.csv")
```

PREPROCESSING

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
var = ['temp','atemp', 'hum', 'windspeed','casual','registered','cnt']
data[var] = scaler.fit_transform(data[var])
```

FINDING LINEAR RELATIONSHIP

```python
import seaborn as sns
sns.heatmap(data.corr(),annot=True)
```

MODEL FITTING

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
X = data[['cnt','temp','atemp']]
X
y = data['registered']
y
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

CHECKING THE ACCURACY

```python
import numpy as np
from sklearn.metrics import r2_score,confusion_matrix,classification_report,mean_squared_error,mean_absolute_error
print("accuracy:",r2_score(y_test,y_pred))
print("mean_absolute_error:",mean_absolute_error(y_test,y_pred))
print("mean_squared_error:",mean_squared_error(y_test,y_pred))
m=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Square Error:",m)
```
